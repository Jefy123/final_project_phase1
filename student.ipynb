```python
df = pd.read_csv('zippedData/tn.movie_budgets.csv.gz')
df
```

```python
#Data cleaning for production budget
# we want to know our return of costs between our net income and and investment and the profit Microsoft studio we are going to make.
#we git rid of money signs in df['production_budget']
df['production_budget'] = df[ 'production_budget'].str.replace('$','')
df
```

```python
#get rid of commas in production budget
df['production_budget'] = df['production_budget'].str.replace(',', '')
df
```

```python
#converting it to a int for calculation
df['production_budget'] = df['production_budget'].astype(int)
df.info()
```

```python
# Data cleaning for ' release date'
df['release_date'] = pd.to_datetime (df['release_date'], infer_datetime_format=True)
df.head(2000)
```

```python
#converting our dates to month 
df['release_month']=pd.DatetimeIndex(df['release_date']).month
df.head()                                        
```

```python
df['worldwide_gross']  = df['worldwide_gross'].str.replace('$', '')
df
```

```python
#converting worldwide gross for analysis
df['worldwide_gross']  = df['worldwide_gross'].str.replace(',', '')
df
    
```

```python
#converting our worldwide to integer....int 64 due to large number of integer
df ['worldwide_gross'] = df['worldwide_gross'].astype('int64')
df.info()
```

```python
#Removing duplicate rows in our dataset.
df = df.drop_duplicates(subset=None, keep='first')
display(df)
```

```python
df.duplicated().value_counts()
```

```python
df[df.duplicated(keep=False)].sort_values(by='id')
```

```python
df['domestic_gross'] = df['domestic_gross'].str.replace('$', '')
df
```

```python
df['domestic_gross'] = df['domestic_gross'].str.replace(',', '')
df
```

```python
df['domestic_gross'] = df['domestic_gross'].astype(int)
df.info()
```

```python
#Data cleaning for value names
df['movie'] = df['movie'].str.lower()
df
```

```python
#Data cleaning to eliminate unused columns of data
df_clean = df.drop(['id', 'domestic_gross'],axis =1)
display(df_clean)
```

```python
# Data analysis 
#My analysis is going to be on the profit made on the selling of the movies.
df['income'] = df['worldwide_gross'] - df['production_budget']
df['income'] = df['income'].astype(int)
df.head()
```

```python
#Analysis of movies based on the return of income the intial budget (production budget) vs income
fig, ax = plt.subplots()
fig.set_size_inches(20,10)
x = df["production_budget"]
y = df["income"]
df =pd.DataFrame({'x':df["production_budget"], 'y': df["income"]})
plt.xlabel('Budget(Dollars)', fontsize = 20, weight ='bold')
plt.ylabel('income (In Billions )', fontsize =20,)
plt.xticks(fontsize=14,rotation=0)
plt.yticks(fontsize=14,rotation=0)
ax.scatter(df.x, df.y, c =np.sign(df.y), cmap="coolwarm")
plt.show()
```
